```python
import numpy as np
import matplotlib.pyplot as plt 
from tensorflow.keras.datasets import mnist
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Laden der MNIST-Daten
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisieren der Daten
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-Hot-Encoding der Labels
y_train_labels = y_train
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Erstellen eines einfachen sequenziellen Modells
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Kompilieren des Modells
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Trainieren des Modells anzeigen
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {accuracy:.2f}")

# Anzeigen der ersten 10 Bilder aus dem Trainingsdatensatz
fig, axes = plt.subplots(1, 10, figsize=(12, 4))
for i, ax in enumerate(axes):
  ax.imshow(X_train[i], cmap='gray')
  ax.set_title(f"Label: {y_train_labels[i]}")
  ax.axis('off')
plt.show()
```
