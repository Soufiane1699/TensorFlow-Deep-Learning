```python
import numpy as np
import tensorflow as tf
```

Daten erstellen

```python
data = np.array([i for i in range(1, 101)]) # Eine einfache Sequenz von 1 bis 100
```

```python
X = []
Y = []
seq_length = 5

for i in range(len(data)-seq_length-1):
  X.append(data[i:i+seq_length])
  Y.append(data[i+seq_length])
```

```python
X = np.array(X)
Y = np.array(Y)
```

Daten skalieren


```python
X = X / 100
Y = Y / 100
```

Modell erstellen


```python
model = tf.keras.models.Sequential([
    tf.keras.layers.SimpleRNN(60, activation='relu', input_shape=[seq_length, 1]),
    tf.keras.layers.Dense(1)
])
```

Modell kompilieren

```python
model.compile(optimizer='adam', loss='mse')
```

Modell trainieren

```python
X = np.expand_dims(X, axis=2) # Die Dimension für die Features hinzufügen
history = model.fit(X, Y, epochs=200)
```

```python
test_data = np.array([96, 97, 98, 99, 100]) / 100.0
test_data = np.expand_dims(test_data, (0, 2)) # Die Batch- und Feature-Dimensionen hinzufügen
```

```python
print(model.predict(test_data) * 100)
```
